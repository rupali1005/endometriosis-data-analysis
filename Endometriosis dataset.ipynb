<H1> Introduction

<H3>Endometriosis is a chronic medical condition where tissue similar to the lining of the uterus, known as the endometrium, grows outside the
uterine cavity. This aberrant growth can occur on various pelvic organs,including the ovaries, fallopian tubes, and the lining of the pelvic cavity.
In rare instances, it may spread beyond the pelvic region.Globally, endometriosis affects approximately 10% of women of reproductive age.

<H2> Exploratory Data Analysis(EDA)

```python
#checking for outliers
plt.figure(figsize=(12,5))
sns.boxplot(data=df[['Age', 'Chronic_Pain_Level', 'BMI']])
plt.title("Checking for outliers")
plt.show()
```

```python
#INTER QUARTILE METHOD
q1 = df['BMI'].quantile(0.25)
q2 = df['BMI'].quantile(0.5)
q3 = df['BMI'].quantile(0.75)
iqr = q3 - q1
```

```python
upper_bound = q3 + 1.5*iqr
upper_bound
```

<h2> Data Visualization</h2>

```python
#Correlation Between Diagnosis & Other Factors
sns.heatmap(df.corr().round(2),annot = True, cmap = 'vlag')
```

<H3> Hormone Level Abnormality and Chronic Pain have the highest correlation with Diagnosis.

<h3> BMI has the weakest correlation with Diagnosis.

```python
#pair plot to show different features in the dataset
plt.figure(figure = (10,5))
sns.pairplot(df, hue = "Diagnosis", diag_kind = "kde")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df,x="BMI",hue="Diagnosis",kde = True,bins = 30, color="orange")
plt.title("KDE plot for BMI Distribution with Diagnosis")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df,x="Age", hue="Diagnosis",kde = True,bins = 30, color="Green")
plt.title("KDE plot of Age Distribution with Diagnosis")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df,x="Chronic_Pain_Level",hue = "Diagnosis",kde = True,bins = 30, color="brown")
plt.title("KDE Plot Distribution of Chronic_Pain_Level with Diagnosis")
plt.ylabel("Frequency")
plt.show()
```

```python
#Proportion of Diagnosed vs Undiagnosed 
df["Diagnosis"].value_counts().plot.pie(autopct="%1.1f%%", colors=["skyblue", "salmon"])
plt.title("Proportion of Diagnosed vs. Undiagnosed")
plt.ylabel("")
plt.show()
```

<h4>A larger undiagnosed slice suggests that a majority of patients have not been diagnosed.

```python
#To Check if BMI and Chronic Pain Level are linked to Diagnosis
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["BMI"], y=df["Chronic_Pain_Level"], hue=df["Diagnosis"], palette="coolwarm")
plt.title("BMI vs. Chronic Pain Level by Diagnosis")
plt.show()
```

```python
#Comparing pain levels between diagnosed and undiagnosed patients
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Diagnosis"], y=df["Chronic_Pain_Level"], palette="vlag")
plt.title("Chronic Pain Level by Diagnosis")
plt.xticks([0, 1], ["Undiagnosed", "Diagnosed"])
plt.show()
```
